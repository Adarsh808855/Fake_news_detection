## Preprocessing Text

#### Creating a function to convert the text in lowercase, remove the extra space, special chr., ulr and links.

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

#### Defining dependent and independent variable as x and y

```python
x = data['text']
y = data['class']
```

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    
    return print("\n\nLR Predicition: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction:{}".format(output_lable(pred_LR[0]),
                                                                                                             output_lable(pred_DT[0]),
                                                                                                             output_lable(pred_GB[0]),
                                                                                                             output_lable(pred_RF[0])))
```

```python
import wikipedia
from textblob import TextBlob
import re

def smart_fake_news_detector(news_text): 
    print("🔍 SMART FAKE NEWS DETECTOR (Fact-Based)")
    
    
    # 1. Extract key claims
    print("\n1️⃣ EXTRACTED CLAIMS:")
    print(f"   '{news_text}'")
    
    # 2. Check credibility indicators
    print("\n2️⃣ CREDIBILITY CHECK:")
    
    credibility_score = 0
    
    # Check for specific numbers/dates
    if re.search(r'\d{4}|\d{1,2}/\d{1,2}', news_text):
        print("   ✅ Contains specific dates/numbers (Good sign)")
        credibility_score += 1
    else:
        print("   ⚠️ No specific dates/numbers (Generic)")
    
    # Check for sources/attribution
    if any(word in news_text.lower() for word in ['reuters', 'ap', 'bbc', 'said', 'reported', 'according']):
        print("   ✅ Has source attribution (Good sign)")
        credibility_score += 1
    else:
        print("   ⚠️ No source mentioned (Questionable)")
    
    # Check for exaggeration
    if any(word in news_text.lower() for word in ['can', 'will', 'definitely', 'always', 'never', 'absolutely']):
        print("   ⚠️ Contains strong language (May be exaggerated)")
        credibility_score -= 0.5
    
    # 3. Wikipedia Verification
    print("\n3️⃣ WIKIPEDIA VERIFICATION:")
    
    # Extract potential entities
    words = news_text.split()
    entities_to_check = []
    
    for i in range(len(words)-1):
        entity = ' '.join(words[i:i+2]).lower()
        if len(entity) > 4:  # Only check meaningful words
            entities_to_check.append(entity)
    
    verified = False
    for entity in entities_to_check[:5]:  # Check first 5 entities
        try:
            result = wikipedia.summary(entity, sentences=1)
            print(f"   ✅ VERIFIED: '{entity}' found in Wikipedia")
            print(f"      → {result[:100]}...")
            verified = True
            credibility_score += 2
            break
        except:
            pass
    
    if not verified:
        print(f"   ❓ Could not verify key claims on Wikipedia")
    
    # 4. Sentiment Analysis
    print("\n4️⃣ LANGUAGE ANALYSIS:")
    blob = TextBlob(news_text)
    sentiment = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    
    if subjectivity > 0.6:
        print(f"   ⚠️ SUBJECTIVE language (opinion-based, not factual)")
        credibility_score -= 1
    else:
        print(f"   ✅ OBJECTIVE language (factual tone)")
    
    # 5. FINAL VERDICT
    print("\n5️⃣ FINAL VERDICT:")
    print("-" * 70)
    
    if credibility_score >= 2:
        print("   ✅ LIKELY REAL NEWS")
        print("   Confidence: HIGH")
    elif credibility_score >= 0:
        print("   ❓ UNCERTAIN / NEEDS VERIFICATION")
        print("   Confidence: MEDIUM")
    else:
        print("   🚨 LIKELY FAKE NEWS")
        print("   Confidence: HIGH")
    
    print(f"   Credibility Score: {credibility_score}/5")
    print("="*70)

```
